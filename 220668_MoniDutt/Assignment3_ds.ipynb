```python
import pandas as pd
import numpy as np
```

```python
wine_data=pd.read_csv('/content/winequality-red.csv')
```

```python
wine_data['bin_quality'] = wine_data['quality'].apply(lambda x: 1 if x >= 7 else 0)
```

```python
new_data=wine_data.drop(['quality'],axis=1)
```

```python
new_data
```

```python
df_1=new_data.drop_duplicates()
df_1
```

```python
df_1.corr()
```

```python
df_2=df_1.describe().T
df_2
```

```python
Q1=df_2['25%']
Q3=df_2['75%']
IQR=(Q3-Q1)
IQR
df_2['lower_outlier']=Q1-1.5*IQR
df_2['upper_outlier']=Q3+1.5*IQR
df_2.T
```

```python
filtered_data = df_1[~((df_1 < (Q1-1.5* IQR)) | (df_1>(Q3 + 1.5*IQR))).any(axis=1)]
filtered_data
```

```python
from sklearn import preprocessing
normalized_arr=preprocessing.normalize(filtered_data)
pd.DataFrame(normalized_arr)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
X=df_1['free sulfur dioxide'].values
y=df_1['bin_quality']
x_train,x_test,y_train,y_test=train_test_split(X.reshape(-1,1),y,test_size=0.3)
classifier=DecisionTreeClassifier()
classifier.fit(x_train,y_train)
y_pred=classifier.predict(x_test)
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
print(f'Confusion_matrix :',cm)
```

```python
TP=cm[0,0]
TN=cm[1,1]
FP=cm[0,1]
FN=cm[1,0]

Accuracy=(TP+TN)/(cm.sum())
print('Accuracy is:',Accuracy*100,'%')
Precision=TP/(TP+FP)
print('Precision is:',Precision)
Recall=TP/(TP+FN)
print('Recall is:',Recall)
F1_score=2*Precision*Recall/(Precision+Recall)
print('F1_score is:',F1_score)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X.reshape(-1,1), y, test_size=0.2, random_state=42)

rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42)

rf_classifier.fit(X_train, y_train)

predictions = rf_classifier.predict(X_test)

cm_rf=confusion_matrix(y_test,predictions)
print(cm_rf)
TP_rf=cm_rf[0,0]
TN_rf=cm_rf[1,1]
FP_rf=cm_rf[0,1]
FN_rf=cm_rf[1,0]

Accuracy_rf=(TP_rf+TN_rf)/(cm_rf.sum())
print('Accuracy is:',Accuracy_rf*100,'%')
Precision_rf=TP_rf/(TP_rf+FP_rf)
print('Precision is:',Precision_rf)
Recall_rf=TP_rf/(TP_rf+FN_rf)
print('Recall is:',Recall_rf)
F1_score_rf=2*Precision_rf*Recall_rf/(Precision_rf+Recall_rf)
print('F1_score is:',F1_score_rf)
```
