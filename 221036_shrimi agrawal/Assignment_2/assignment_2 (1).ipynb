```python
df = pds.read_csv("Social_Network_Ads (1).csv")
df.head()
```

```python
X = df['Age'].values
y = df['Purchased'].values
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.30)
print(y_test.size)
```

```python
def normalize(X):
  return X - X.mean()
```

```python
def standardize(X):
  return (X - X.mean())/(X.std())
```

```python
# method to make predictions
def predict(X,b0,b1):
  return np.array([1/(1 + exp(-1*b0 + -1*b1*x)) for x in X])
#to make a logicstic regression model
#for raw data
def logistic_regression_1(X,Y):
  #initialise values
  b0 = 0;
  b1 = 0;
  l = 0.0001;
  epochs = 1000;
  for epoch in range(epochs):
    y_pred = predict(X,b0,b1)
    D_b0 = -2*sum((Y - y_pred)*y_pred*(1-y_pred))
    D_b1 = -2*sum(X*(Y - y_pred)*y_pred*(1-y_pred))
    b0 =b0 - l*D_b0;
    b1 = b1 - l*D_b1;
  return b0,b1
#train model
b0_1,b1_1 = logistic_regression_1(X_train,y_train)
# make predictions
y_pred = predict(X_test,b0_1,b1_1)
y_pred = [1 if p >= 0.5 else 0 for p in y_pred]
#accuracy
accuracy =0
for i in range(len(y_pred)):
  if y_pred[i] == y_test[i]:
    accuracy += 1;
print(f"accuracy_1 = {accuracy/len(y_pred)}")
```

```python
#for normalised data
def logistic_regression_2(X,Y):
  #initialise values
  b0 = 0;
  b1 = 0;
  l = 0.0001;
  epochs = 1000;
  X1 = normalize(X)
  for epoch in range(epochs):
    y_pred = predict(X1,b0,b1)
    D_b0 = -2*sum((Y - y_pred)*y_pred*(1-y_pred))
    D_b1 = -2*sum(X1*(Y - y_pred)*y_pred*(1-y_pred))
    b0 =b0 - l*D_b0;
    b1 = b1 - l*D_b1;
  return b0,b1
#train model
b0_2,b1_2 = logistic_regression_1(X_train,y_train)
# make predictions
X_normalize_test = normalize(X_test)
y_pred = predict(X_normalize_test,b0_2,b1_2)
y_pred = [1 if p >= 0.5 else 0 for p in y_pred]
#accuracy
accuracy =0
for i in range(len(y_pred)):
  if y_pred[i] == y_test[i]:
    accuracy += 1;
print(f"accuracy_2 = {accuracy/len(y_pred)}")
```

```python
#for standardized data
def logistic_regression_3(X,Y):
  #initialise values
  b0 = 0;
  b1 = 0;
  l = 0.0001;
  epochs = 1000;
  X1 = standardize(X)
  for epoch in range(epochs):
    y_pred = predict(X1,b0,b1)
    D_b0 = -2*sum((Y - y_pred)*y_pred*(1-y_pred))
    D_b1 = -2*sum(X1*(Y - y_pred)*y_pred*(1-y_pred))
    b0 =b0 - l*D_b0;
    b1 = b1 - l*D_b1;
  return b0,b1
#train model
b0_3,b1_3 = logistic_regression_1(X_train,y_train)
# make predictions
X_standardize_test = standardize(X_test)
y_pred = predict(X_standardize_test,b0_3,b1_3)
y_pred = [1 if p >= 0.5 else 0 for p in y_pred]
#accuracy
accuracy =0
for i in range(len(y_pred)):
  if y_pred[i] == y_test[i]:
    accuracy += 1;
print(f"accuracy_3 = {accuracy/len(y_pred)}")
```
