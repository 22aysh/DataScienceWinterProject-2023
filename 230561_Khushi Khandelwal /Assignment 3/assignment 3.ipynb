```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('winequality-red.csv')
```

```python
df['good_quality']=(df['quality']>=7).astype(int)
df=df.drop('quality',axis=1)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.utils import shuffle
```

```python
df=shuffle(df)
```

```python
X = df.drop('good_quality', axis=1)
y = df['good_quality']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
df.corr()
```

```python
df.describe()
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, f1_score
```

```python
dtc = DecisionTreeClassifier(random_state=42)
```

```python
dtc.fit(X_train, y_train)
```

```python
dt_predictions = dtc.predict(X_test)
```

```python
dt_accuracy = accuracy_score(y_test, dt_predictions)
dt_f1_score = f1_score(y_test, dt_predictions)
```

```python
print("Desicion Tree Classifier : ")
print(f"Accuracy: {dt_accuracy:.2f}")
print(f"F1 Score: {dt_f1_score:.2f}")
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc=RandomForestClassifier(random_state=42)
```

```python
rfc.fit(X_train, y_train)
```

```python
rf_predictions = rfc.predict(X_test)
```

```python
rf_accuracy = accuracy_score(y_test, rf_predictions)
rf_f1_score = f1_score(y_test, rf_predictions)
```

```python
print("Random Forest Classifier : ")
print(f"Accuracy: {rf_accuracy:.2f}")
print(f"F1 Score: {rf_f1_score:.2f}")
```
